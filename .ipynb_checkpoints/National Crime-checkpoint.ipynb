Pulled crime data from [here](https://www.ucrdatatool.gov/Search/Crime/State/StateCrime.cfm). Removed the titles and footnotes, and put each year into one CSV.

```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from ipywidgets import interactive

def fullprint(df):
    with pd.option_context('display.max_rows', None, 'display.max_columns', None):
        display(df)
        
agg = pd.read_csv('/Users/LOML/Documents/GitHub/National-Crime/Aggregate2.csv')

# replace spaces and decimal points in the column names.
agg.columns = agg.columns.str.replace('\s+', '_')
agg.columns = agg.columns.str.replace('.', '')
agg.columns = agg.columns.str.replace('/', '')
```

```python
fullprint(agg.head())
```

Used [this](https://blog.ouseful.info/2016/12/29/simple-view-controls-for-pandas-dataframes-using-ipython-widgets/) as a guide for creating the widget below.

```python
vlist1 = ['Violent_crime_total','Murder_and_nonnegligent_Manslaughter','Motor_vehicle_theft']
vlist2 = ['Legacy_rape_1','Robbery','Aggravated_assault']
vlist3 = ['Property_crime_total','Burglary','Larceny-theft']
    
items = ['All']+sorted(agg['State'].unique().tolist())

sns.set_style("dark")

def view(x=''):
    x==agg['State']
    if x!='All':
        fig, ax =plt.subplots(1,1)
        fig.subplots_adjust(top=1, bottom=0, left=0, right=1.8,hspace=.2, wspace=.4)
        an = sns.pointplot(x="Year", y="Population", hue="State", data=agg[agg['State']==x], dodge=True, markers="None")
        an.set(xticklabels=[])
        an.legend_.remove()
        fig, ax =plt.subplots(1,3)
        fig.subplots_adjust(top=.5, bottom=0, left=0, right=2,hspace=.2, wspace=.4)
        for i,col in enumerate(vlist1):
            an = sns.pointplot(x="Year", y=col, hue="State", data=agg[agg['State']==x], ax=ax[i-1],dodge=True, markers="None")
            an.set(xticklabels=[])
            an.legend_.remove()
        fig, ax =plt.subplots(1,3)
        fig.subplots_adjust(top=.5, bottom=0, left=0, right=2,hspace=.2, wspace=.4)
        for i,col in enumerate(vlist2):
            an = sns.pointplot(x="Year", y=col, hue="State", data=agg[agg['State']==x], ax=ax[i-1],dodge=True, markers="None")
            an.set(xticklabels=[])
            an.legend_.remove()
        fig, ax =plt.subplots(1,3)
        fig.subplots_adjust(top=.5, bottom=0, left=0, right=2,hspace=.2, wspace=.4)
        for i,col in enumerate(vlist3):
            an = sns.pointplot(x="Year", y=col, hue="State", data=agg[agg['State']==x], ax=ax[i-1],dodge=True, markers="None")
            an.set(xticklabels=[])
            an.legend_.remove()

w = widgets.Select(options=items)
interactive(view, x=w)
```
